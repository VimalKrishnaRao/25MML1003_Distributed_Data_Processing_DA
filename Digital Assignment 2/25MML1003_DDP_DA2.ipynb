## (a): Theory – Transformations vs Actions in Spark

## Introduction
In Apache Spark, operations on data are broadly classified into transformations and actions. These concepts are fundamental to understanding how Spark processes data efficiently using lazy evaluation and execution planning.

## Transformations in Spark
Transformations are operations that create a new DataFrame or RDD from an existing one without immediately executing the computation. They are lazy in nature.

### Examples of Transformations
- map() – Applies a function to each element  
- filter() – Selects elements based on a condition  
- select() – Chooses specific columns from a DataFrame  
- groupBy() – Groups data based on column values  

### Key Characteristics
- Lazy execution  
- Returns a new dataset  
- Builds a Directed Acyclic Graph (DAG)  
- Optimized before execution  

## Actions in Spark
Actions are operations that trigger the execution of transformations and return a result to the driver program or write data to storage.

### Examples of Actions
- count() – Returns the number of elements  
- collect() – Retrieves all data to the driver  
- show() – Displays data in tabular format  
- write() – Stores data to external systems  

### Key Characteristics
- Eager execution  
- Produces output or result  
- Triggers DAG execution  
- Executes all preceding transformations  

## Differences Between Transformations and Actions

| Feature | Transformations | Actions |
|--------|----------------|--------|
| Execution | Lazy | Eager |
| Output | New DataFrame/RDD | Final result/value |
| Role | Build execution plan | Trigger execution |
| Example | map(), filter() | count(), collect() |

## Significance in Execution Planning

### Lazy Evaluation and Optimization
Transformations are not executed immediately. Spark builds a logical execution plan (DAG) and optimizes it before execution.

### DAG Formation
Each transformation adds a step in the DAG. When an action is called, Spark executes the DAG efficiently.

### Reduced Data Movement
Spark minimizes data shuffling across nodes, improving performance.

### Fault Tolerance
Spark can recompute lost data using lineage information.

### Efficient Resource Utilization
Execution is delayed until necessary, ensuring better use of system resources.

## Conclusion
Transformations and actions are core concepts in Spark’s execution model. Transformations define what needs to be done, while actions trigger the computation. This separation enables efficient, scalable, and optimized data processing.

```python
# Import required modules
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count
```

```python
# Create Spark Session

spark = SparkSession.builder \
    .appName("PySpark Transformations Example") \
    .getOrCreate()
```

SparkSession is the entry point for all PySpark operations

```python
# Manually Create Dataset

data = [
    (1, "Alice", 25, "HR", 25000),
    (2, "Bob", 30, "IT", 40000),
    (3, "Charlie", 28, "IT", 32000),
    (4, "David", 35, "Finance", 28000),
    (5, "Eve", 29, "HR", 50000)
]

columns = ["ID", "Name", "Age", "Department", "Salary"]

df = spark.createDataFrame(data, columns)
```

```python
# Select Specific Columns

selected_df = df.select("Name", "Department", "Salary")
selected_df.show()
```


* select() is used to retrieve only required columns
* Reduces unnecessary data processing

```python
# Apply Filter Condition

filtered_df = df.filter(col("Salary") > 30000)
filtered_df.show()
```

* Filters employees with salary greater than 30,000
* col() helps reference column names

```python
# Group By and Count

grouped_df = df.groupBy("Department").agg(count("*").alias("Employee_Count"))
grouped_df.show()
```

* groupBy() groups data based on department
* count() counts number of employees in each group
* alias() renames the column

```python
# Sort by Salary

df.orderBy(col("Salary").desc()).show()
```

```python
# Display First Record

df.first()
```

```python
# Count Total Records

df.count()
```

This experiment demonstrates how PySpark performs efficient data transformations on structured datasets. Using operations like selection, filtering, and grouping, we can analyze large datasets easily in a distributed environment.
