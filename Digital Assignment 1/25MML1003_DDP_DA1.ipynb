## (a): Distributed Data Processing and Apache Spark

Distributed Data Processing is a method of handling large-scale data by dividing it into smaller parts and processing them across multiple machines (nodes) in a cluster. This approach improves efficiency, scalability, and fault tolerance compared to traditional single-machine processing systems.

Frameworks such as Apache Spark and Hadoop enable distributed computation by managing data distribution and parallel execution.

## Key Concepts of Distributed Data Processing

### Parallel Processing
Data is processed simultaneously across multiple nodes, reducing execution time.

### Scalability
Systems can scale horizontally by adding more machines to handle increasing data volumes.

### Fault Tolerance
Failure of one node does not affect the entire system; data can be recovered using replication or recomputation.

### Data Partitioning
Large datasets are divided into smaller partitions and distributed across cluster nodes.

### Cluster Computing
Multiple interconnected systems work together as a single unit.

## Limitations of Traditional Data Processing Systems

Traditional systems like Hadoop MapReduce rely heavily on disk-based processing:

- Intermediate results are written to disk after each step  
- High disk I/O increases latency  
- Inefficient for iterative tasks like machine learning  
- Not suitable for real-time or interactive processing  

## How Apache Spark Improves Performance

Apache Spark is a fast, in-memory distributed data processing framework designed to overcome the limitations of traditional systems.

### In-Memory Computation
- Stores intermediate data in RAM instead of disk  
- Reduces disk I/O operations  
- Provides 10–100x faster processing than MapReduce  
- Ideal for iterative and real-time applications  

### Resilient Distributed Datasets (RDDs)
- Fault-tolerant distributed data structures  
- Support transformations (map, filter) and actions (count, collect)  
- Automatically recover lost data using lineage  

### Lazy Evaluation
- Operations are not executed immediately  
- Builds an optimized execution plan (DAG) before running  

### DAG Execution Engine
- Uses Directed Acyclic Graph (DAG) for scheduling  
- Optimizes execution and minimizes unnecessary computations  

### Unified Analytics Engine
Supports multiple workloads:
- Batch processing  
- Stream processing  
- Machine learning (MLlib)  
- Graph processing (GraphX)  

### Reduced Disk Dependency
- Writes to disk only when necessary  
- Improves efficiency and speed  

## Role of In-Memory Computation

In-memory computation is the core reason for Spark’s performance:

- Data is cached in RAM for reuse  
- Eliminates repeated disk access  
- Speeds up iterative algorithms  
- Enables real-time data analysis  

## Conclusion

Distributed data processing enables efficient handling of large datasets by leveraging multiple machines. Apache Spark enhances this process through in-memory computation, optimized execution using DAG, and reduced reliance on disk operations. As a result, Spark provides significantly faster and more flexible data processing compared to traditional systems.

# (b): PySpark Program: Employee DataFrame Operations

### Aim

The aim of this experiment is to demonstrate how to use PySpark to:

* Create a DataFrame with employee records
* Define a schema (ID, Name, Age, Salary)
* Display the structure (schema) of the DataFrame
* Show all records present in the DataFrame
* Filter employees based on a salary condition (Salary > 30,000)

### Objective
* To understand how PySpark handles structured data
* To perform basic DataFrame operations such as creation, visualization, and filtering
* To learn schema inspection and transformation techniques

```python
!pip install pyspark
```

```python
# Import required libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
# Create Spark Session
spark = SparkSession.builder \
    .appName("Employee DataFrame Example") \
    .getOrCreate()
```



* SparkSession is the entry point to PySpark
* It allows us to create DataFrames and execute operations

```python
# Create DataFrame
# Sample data
data = [
    (1, "Alice", 25, 25000),
    (2, "Bob", 30, 40000),
    (3, "Charlie", 28, 32000),
    (4, "David", 35, 28000),
    (5, "Eve", 29, 50000)
]

# Column names
columns = ["ID", "Name", "Age", "Salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)
```


* Data is created as a list of tuples
* Columns define schema
* DataFrame is created using createDataFrame()

```python
# Display Schema
# Print schema
df.printSchema()
```

* Shows structure of DataFrame
* Displays column names, data types, and nullability

```python
# Show all records
df.show()
```

* Displays all rows in tabular format
* Useful for quick inspection of data

```python
# Filter Employees with Salary > 30,000
# Filter employees

filtered_df = df.filter(col("Salary") > 30000)

# Show filtered results
filtered_df.show()
```

* filter() is used to apply condition
* col("Salary") > 30000 selects only high-salary employees

```python
# Select Specific Columns

df.select("Name", "Salary").show()
```

```python
# Sort by Salary

df.orderBy(col("Salary").desc()).show()
```

```python
# Add New Column

df.withColumn("Bonus", col("Salary") * 0.1).show()
```

### Result

This experiment demonstrates how PySpark simplifies big data processing using DataFrames. It provides efficient operations such as filtering, transformation, and schema handling, making it a powerful tool for distributed data processing.
